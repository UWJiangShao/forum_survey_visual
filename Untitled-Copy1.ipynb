```python
file_path1 = '/Users/tigershao/Dropbox (UFL)/Forum Slides Making Project/data/admin data/all_year/hedis_star_all.csv'
file_path2 = '/Users/tigershao/Dropbox (UFL)/Forum Slides Making Project/data/admin data/all_year/hedis_sk_all.csv'
file_path3 = '/Users/tigershao/Dropbox (UFL)/Forum Slides Making Project/data/admin data/all_year/hedis_sp_all.csv'
file_path4 = '/Users/tigershao/Dropbox (UFL)/Forum Slides Making Project/data/admin data/all_year/national_level_all.csv'
hedis_star_all = pd.read_csv(file_path1)
hedis_sk_all = pd.read_csv(file_path2)
hedis_sp_all = pd.read_csv(file_path3)
national_level_all = pd.read_csv(file_path4)
hedis_star_all.head()
national_level_all.head(10)
```

```python
hedis_star_all.head(10)
```

```python
##########################
#'Chlamydia Screen (CHL) from 2018 to 2022'
##########################


filtered_data = hedis_star_all[(hedis_star_all['MeasureNameAbbreviation'] == 'CHL') &
                               (hedis_star_all['PopulationName'] == 'ALL') &
                               (hedis_star_all['SubmeasureName'] == 'Total')]

filtered_national = national_level_all[
    (national_level_all['MeasureNameAbbreviation'] == 'CHL') &
    (national_level_all['MeasureName'].str.contains('Chlamydia Screening in Women.*Total', case=False, na=False))
]

plt.rcParams.update({'font.size': 25})

plt.figure(figsize=(16, 7), facecolor='white')

sns.lineplot(data=filtered_data, x='year', y='rate', marker='o', color='black', label = 'State Overall') 
sns.lineplot(data=filtered_national, x='year', y='X50thPercentile', marker='o', color='#FF9933', label='50th NCQA', linewidth=3.5)
sns.lineplot(data=filtered_national, x='year', y='X25thPercentile', marker='o', color='#FF0000', label='25th NCQA', linewidth=3.5)

plt.title('Chlamydia Screen (CHL) from 2018 to 2022')
plt.xlabel('')
plt.ylabel('Rate')
plt.grid(False)

# Ensure every year is shown on the x-axis, no .0 or .5
plt.xticks(filtered_data['year'].unique())  
plt.tight_layout()  # Adjusts subplot params so that the subplot(s) fits in to the figure area
plt.ylim(43, 63)
plt.show()
```

```python
filtered_data['year'] = filtered_data['year'].astype(str)
filtered_national['year'] = filtered_national['year'].astype(str)

plt.rcParams.update({'font.size': 25})

plt.figure(figsize=(16, 7), facecolor='white')

sns.barplot(data=filtered_data, x='year', y='rate', color='grey', label='State Overall')

sns.lineplot(data=filtered_national, x='year', y='X50thPercentile', marker='o', color='#FF9933', label='50th NCQA', linewidth=3.5)
sns.lineplot(data=filtered_national, x='year', y='X25thPercentile', marker='o', color='#FF0000', label='25th NCQA', linewidth=3.5)

plt.title('Chlamydia Screen (CHL) from 2018 to 2022')
plt.xlabel('')
plt.ylabel('Rate')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.ylim(42, 65)
plt.legend()
plt.show()
```
